# why not similarity?

```python
gene = cat.read_upload(gene_file)
metabo = cat.read_upload(metabo_file)
design = cat.read_upload(design_file)

processed_gene = cat.repeat_aggregation_max(gene, design)
processed_metabo = cat.repeat_aggregation_max(metabo, design)

t = cat.get_target(target, processed_metabo)
```

```python
spearman = cat.compute_spearman(processed_gene, t)
pearson = cat.compute_pearson(processed_gene, t)
granger = cat.compute_granger(processed_gene, t)
fc = cat.fc_comp()
```

```python
data = cat.merge_dataframes([processed_gene, spearman, pearson, granger, fc])
data = cat.scale_column(data, 'Pearson')
data = cat.scale_column(data, 'Spearman')
data = data.fillna(0)
```

```python
cat.plot_line(t,'Capsaicin')
```

```python
cat.plot_line_heatmap(data, 'Cc72_00034724-RA', cmap='vlag')
```
